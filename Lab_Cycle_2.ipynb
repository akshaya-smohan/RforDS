Create a scatterplot of the mpg and disp variables in the mtcars

**Program 4 :**

Create a histogram of the miles per gallon (mpg) in the mtcars dataset.
Use different shades of blue to represent the frequency of each bin.
Add appropriate title and labels to the plot. Calculate and display the
mean and standard deviation of mpg on the plot.

```r
# Load the mtcars dataset
data(mtcars)

# Calculate mean and standard deviation of mpg
mean_mpg <- mean(mtcars$mpg)
sd_mpg <- sd(mtcars$mpg)

# Create a color palette of blue shades
blue_shades <- colorRampPalette(c("lightblue", "blue"))(10)

# Create the histogram
hist_info <- hist(
  mtcars$mpg,
  breaks = 10,
  col = blue_shades,
  main = "Histogram of Miles Per Gallon (mpg)",
  xlab = "Miles Per Gallon (mpg)",
  ylab = "Frequency",
  border = "white"
)

# Add vertical lines for mean and 1 standard deviation
abline(v = mean_mpg, col = "red", lwd = 2, lty = 2)
abline(v = mean_mpg + sd_mpg, col = "darkgreen", lwd = 2, lty = 3)
abline(v = mean_mpg - sd_mpg, col = "darkgreen", lwd = 2, lty = 3)

# Add text displaying mean and standard deviation
text(
  x = mean_mpg,
  y = max(hist_info$counts) * 0.9,
  labels = paste("Mean =", round(mean_mpg, 2)),
  col = "red",
  pos = 4
)
text(
  x = mean_mpg,
  y = max(hist_info$counts) * 0.8,
  labels = paste("SD =", round(sd_mpg, 2)),
  col = "darkgreen",
  pos = 4
)

# Save as a high-resolution PNG image
png("mtcars_histogram.png", width = 2000, height = 1500, res = 300)

# Recreate the plot inside the PNG device
hist_info <- hist(
  mtcars$mpg,
  breaks = 10,
  col = blue_shades,
  main = "Histogram of Miles Per Gallon (mpg)",
  xlab = "Miles Per Gallon (mpg)",
  ylab = "Frequency",
  border = "white"
)
abline(v = mean_mpg, col = "red", lwd = 2, lty = 2)
abline(v = mean_mpg + sd_mpg, col = "darkgreen", lwd = 2, lty = 3)
abline(v = mean_mpg - sd_mpg, col = "darkgreen", lwd = 2, lty = 3)
text(
  x = mean_mpg,
  y = max(hist_info$counts) * 0.9,
  labels = paste("Mean =", round(mean_mpg, 2)),
  col = "red",
  pos = 4
)
text(
  x = mean_mpg,
  y = max(hist_info$counts) * 0.8,
  labels = paste("SD =", round(sd_mpg, 2)),
  col = "darkgreen",
  pos = 4
)

# Close the device to save the image
dev.off()
```

**Program 5 :**

Create a box plot of the horsepower (hp) in the mtcars dataset. Use
different shapes to represent the number of gears (gear). Add
appropriate title, labels, and legend to the plot. Identify and label any
outliers on the plot.

```r
# Load the mtcars dataset
data(mtcars)

# Convert 'gear' to a factor for grouping
mtcars$gear <- factor(mtcars$gear)

# Create the boxplot
box_info <- boxplot(
  hp ~ gear,
  data = mtcars,
  main = "Box Plot of Horsepower by Number of Gears",
  xlab = "Number of Gears",
  ylab = "Horsepower (hp)",
  col = c("lavender", "orange", "pink"),
  pch = 19,
  border = "darkgray"
)

# Check and label outliers only if they exist
if (length(box_info$out) > 0) {
  for (i in seq_along(box_info$out)) {
    text(
      x = box_info$group[i] + 0.1,
      y = box_info$out[i],
      labels = round(box_info$out[i], 1),
      col = "red",
      cex = 0.8,
      pos = 4
    )
  }
}

# Add legend
legend(
  "topright",
  legend = levels(mtcars$gear),
  fill = c("lavender", "orange", "pink"),
  title = "Number of Gears"
)

# Save as a high-resolution PNG
png("mtcars_boxplot_hp.png", width = 2000, height = 1500, res = 300)

# Recreate inside PNG device
box_info <- boxplot(
  hp ~ gear,
  data = mtcars,
  main = "Box Plot of Horsepower by Number of Gears",
  xlab = "Number of Gears",
  ylab = "Horsepower (hp)",
  col = c("lavender", "orange", "pink"),
  pch = 19,
  border = "darkgray"
)

# Label outliers safely
if (length(box_info$out) > 0) {
  for (i in seq_along(box_info$out)) {
    text(
      x = box_info$group[i] + 0.1,
      y = box_info$out[i],
      labels = round(box_info$out[i], 1),
      col = "red",
      cex = 0.8,
      pos = 4
    )
  }
}
```

**Program 6 :**

Create a scatter plot of the displacement (disp) versus the weight (wt)
in the mtcars dataset. Use different colors and sizes to represent the
number of carburetors (carb). Add appropriate title, labels, and legend
to the plot. Add a smooth line to show the trend of the relationship.

```r
# Load the mtcars dataset
data(mtcars)

# Convert 'carb' to a factor
mtcars$carb <- factor(mtcars$carb)

# Define colors and sizes based on carburetors
colors <- c("red", "blue", "green", "orange", "purple", "brown")[as.numeric(mtcars$carb)]
sizes <- 1.5 + as.numeric(mtcars$carb) * 0.5

# Create the scatter plot
plot(
  mtcars$disp,
  mtcars$wt,
  col = colors,
  pch = 19,
  cex = sizes,
  main = "Scatter Plot of Displacement vs Weight",
  xlab = "Displacement (cu. in.)",
  ylab = "Weight (1000 lbs)"
)

# Add a smooth trend line
lines(lowess(mtcars$disp, mtcars$wt), col = "black", lwd = 2)

# Add a neatly arranged legend (below the plot)
legend(
  "topleft",
  legend = paste(levels(mtcars$carb), "Carburetors"),
  col = c("red", "blue", "green", "orange", "purple", "brown")[1:length(levels(mtcars$carb))],
  pch = 19,
  pt.cex = 1.5 + as.numeric(levels(mtcars$carb)) * 0.5,
  title = "Number of Carburetors"
)

# Save as a high-resolution PNG
png("mtcars_disp_wt_scatter_neat.png", width = 2000, height = 1500, res = 300)

# Recreate the plot inside the PNG device
plot(
  mtcars$disp,
  mtcars$wt,
  col = colors,
  pch = 19,
  cex = sizes,
  main = "Scatter Plot of Displacement vs Weight",
  xlab = "Displacement (cu. in.)",
  ylab = "Weight (1000 lbs)"
)
lines(lowess(mtcars$disp, mtcars$wt), col = "black", lwd = 2)
legend(
  "topleft",
  legend = paste(levels(mtcars$carb), "Carburetors"),
  col = c("red", "blue", "green", "orange", "purple", "brown")[1:length(levels(mtcars$carb))],
  pch = 19,
  pt.cex = 1.5 + as.numeric(levels(mtcars$carb)) * 0.5,
  title = "Number of Carburetors"
)

dev.off()
```

**Program 7:**

Develop an R program to create a time series plot using real-world
data. (https://www.kaggle.com/datasets/niketchauhan/covid-19-time-
series-data)

```r
# ================================
#  Install and load libraries
# ================================
if(!require(tidyverse)) install.packages("tidyverse", dependencies=TRUE)
if(!require(lubridate)) install.packages("lubridate", dependencies=TRUE)
if(!require(scales)) install.packages("scales", dependencies=TRUE)

library(tidyverse)
library(lubridate)
library(scales)

# ================================
#  Read your uploaded dataset
# ================================
# Youve uploaded: time-series-19-covid-combined.csv
covid <- read_csv("time-series-19-covid-combined.csv")

# View first few rows
head(covid)

# ================================
#  Data cleaning
# ================================
# Convert Date column to Date format
covid <- covid %>%
  mutate(Date = as.Date(Date)) %>%
  rename(
    Country = `Country/Region`
  )

# Check available columns
glimpse(covid)

# ================================
#  Select country of interest
# ================================
country_of_interest <- "India"

covid_country <- covid %>%
  filter(Country == country_of_interest) %>%
  group_by(Date) %>%
  summarise(
    Confirmed = sum(Confirmed, na.rm = TRUE),
    Deaths = sum(Deaths, na.rm = TRUE),
    Recovered = sum(Recovered, na.rm = TRUE)
  )

# ================================
#  Plot time series with smooth line
# ================================
ggplot(covid_country, aes(x = Date, y = Confirmed)) +
  geom_line(color = "steelblue", size = 1) +
  geom_point(color = "steelblue", size = 1.2, alpha = 0.7) +
  geom_smooth(method = "loess", span = 0.2, se = TRUE, color = "darkred") +
  scale_x_date(date_labels = "%b %Y", date_breaks = "2 months") +
  scale_y_continuous(labels = comma) +
  labs(
    title = paste("COVID-19 Confirmed Cases in", country_of_interest),
    subtitle = "Time Series of Confirmed Cases (with Smooth Trend Line)",
    x = "Date",
    y = "Confirmed Cases",
    caption = "Source: Kaggle COVID-19 Time Series Data"
  ) +
  theme_minimal(base_size = 14) +
  theme(
    axis.text.x = element_text(angle = 45, hjust = 1),
    plot.title = element_text(face = "bold", hjust = 0.5),
    plot.subtitle = element_text(hjust = 0.5)
  )
```

```r
# Load necessary libraries
library(ggplot2)
library(dplyr)
library(psych)
library(ggpubr)
library(corrplot)

# Read the Titanic dataset (replace 'titanic.csv' with your filename if needed)
titanic <- read.csv('titanic.csv', stringsAsFactors = TRUE)

# a) Histogram of Number of Parents and Children aboard (parch)
ggplot(titanic, aes(x = parch)) +
  geom_histogram(binwidth = 1, fill = "skyblue", color = "black") +
  labs(title = "Histogram of Number of Parents and Children (parch)",
       x = "Number of Parents/Children Aboard (parch)",
       y = "Count") +
  theme_minimal()

# b) Detailed EDA: Advanced statistical analysis for survival factors

# 1. Statistical Summary
summary(titanic)
describe(titanic)

# 2. Survival rates by different factors
survival_by_class <- titanic %>%
  group_by(Pclass) %>%
  summarise(SurvivalRate = mean(Survived, na.rm = TRUE))

survival_by_sex <- titanic %>%
  group_by(Sex) %>%
  summarise(SurvivalRate = mean(Survived, na.rm = TRUE))

survival_by_embark <- titanic %>%
  group_by(Embarked) %>%
  summarise(SurvivalRate = mean(Survived, na.rm = TRUE))

# Print summaries
print(survival_by_class)
print(survival_by_sex)
print(survival_by_embark)

# 3. Chi-squared test: Survival vs. Pclass, Sex
chisq.test(table(titanic$Survived, titanic$Pclass))
chisq.test(table(titanic$Survived, titanic$Sex))

# 4. Age: Check if age distributions differ by survival
t.test(titanic$Age ~ titanic$Survived, na.rm = TRUE)

# 5. Correlation matrix (numeric variables only)
numeric_vars <- titanic %>%
  select_if(is.numeric) %>%
  na.omit()
corrplot(cor(numeric_vars), method = "circle")

# 6. Logistic Regression (optional advanced step)
titanic_clean <- titanic %>% mutate(Sex = as.numeric(Sex=="male"))
model <- glm(Survived ~ Pclass + Sex + Age + SibSp + Parch + Fare, data = titanic_clean, family = "binomial")
summary(model)

# c) Customized box plot: Age distribution by survival

titanic_surv <- titanic %>% filter(!is.na(Age))
ggplot(titanic_surv, aes(x = factor(Survived, labels = c("Not Survived", "Survived")), y = Age, fill = factor(Survived))) +
  geom_boxplot(width=0.5, outlier.color="red", outlier.shape=8, notch=TRUE) +
  scale_fill_manual(values = c("#F8766D", "#00BA38")) +
  labs(title = "Age Distribution by Survival Status",
       x = "Survival Status",
       y = "Age") +
  theme_minimal() +
  theme(legend.position = "none")
```
